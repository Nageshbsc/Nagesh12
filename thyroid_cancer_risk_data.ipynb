```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("thyroid_cancer_risk_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe() 
```

```python
df.isna().sum()
```

```python
df['Age'].plot(kind = "hist", edgecolor = "black")
plt.xlabel("Age")
plt.title("Age vs Country")
plt.show()
```

```python
df["Gender"].value_counts().plot(kind = 'bar', color = ['blue', 'orange'])
plt.title("Gender vs TSH_Level")
plt.grid(alpha = 0.8)
plt.show()
```

```python
df.groupby("Family_History")['Ethnicity'].value_counts()
```

```python
df['Ethnicity'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%' )
plt.title("Pie chart of Thyroid cancert")
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Family_History', hue='Ethnicity')
plt.title("Ethnicity Distribution by Family History")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
